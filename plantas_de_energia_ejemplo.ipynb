```python
import plotly.express as px

# Calcular la capacidad total de generaciÃ³n por paÃ­s y ordenar de mayor a menor
capacidad_por_pais = df_americano.groupby('country')['capacity_mw'].sum().reset_index()
capacidad_por_pais = capacidad_por_pais.sort_values(by='capacity_mw', ascending=True)

# Crear un grÃ¡fico de barras horizontales ordenado de mayor a menor capacidad de generaciÃ³n
fig = px.bar(capacidad_por_pais, x='capacity_mw', y='country', orientation='h',
             title='Capacidad de GeneraciÃ³n de EnergÃ­a por PaÃ­s en AmÃ©rica',
             labels={'capacity_mw': 'Capacidad Total (MW)', 'country': 'PaÃ­s'},
             width=800, height=600)  # Aumentar el tamaÃ±o del grÃ¡fico

# Mostrar el grÃ¡fico interactivo
fig.show()
```

```python
import plotly.express as px
# Calcular la capacidad total de generaciÃ³n en AmÃ©rica
capacidad_total_americana = df_americano['capacity_mw'].sum()

# Calcular el porcentaje de capacidad de generaciÃ³n por paÃ­s
capacidad_por_pais['capacidad_porcentaje'] = capacidad_por_pais['capacity_mw'] / capacidad_total_americana * 100

# Agrupar paÃ­ses cuya capacidad de generaciÃ³n sea menor al 1% del total
umbral = 1  # Umbral del 1%
otros = capacidad_por_pais[capacidad_por_pais['capacidad_porcentaje'] < umbral]
otros_total = otros['capacity_mw'].sum()
otros_porcentaje = otros_total / capacidad_total_americana * 100

# Crear una fila para 'Otros' con la capacidad combinada
otros_row = pd.DataFrame({'country': ['Otros'], 'capacity_mw': [otros_total], 'capacidad_porcentaje': [otros_porcentaje]})

# Filtrar los paÃ­ses que superan el umbral
capacidad_por_pais = capacidad_por_pais[capacidad_por_pais['capacidad_porcentaje'] >= umbral]

# Concatenar 'Otros' con los paÃ­ses que superan el umbral
capacidad_por_pais = pd.concat([capacidad_por_pais, otros_row])

# Crear un grÃ¡fico de pastel
fig = px.pie(capacidad_por_pais, values='capacity_mw', names='country',
             title='DistribuciÃ³n de Capacidad de GeneraciÃ³n de EnergÃ­a por PaÃ­s en AmÃ©rica',
             hole=0.4,  # TamaÃ±o del agujero en el centro del grÃ¡fico (0 a 1)
             )

# Mostrar el grÃ¡fico interactivo
fig.show()
```

```python
import plotly.express as px

# Agrupar por paÃ­s y tipo de combustible, y calcular la capacidad total de generaciÃ³n
capacidad_por_tipo_combustible = df_americano.groupby(['country', 'primary_fuel'])['capacity_mw'].sum().reset_index()

# Calcular la capacidad total de generaciÃ³n por paÃ­s
capacidad_total_por_pais = capacidad_por_tipo_combustible.groupby('country')['capacity_mw'].sum().reset_index()

# Ordenar los paÃ­ses por capacidad total de generaciÃ³n de mayor a menor
capacidad_total_por_pais = capacidad_total_por_pais.sort_values(by='capacity_mw', ascending=False)

# Filtrar los datos para incluir solo los paÃ­ses en orden descendente de capacidad de generaciÃ³n
paises_ordenados = capacidad_total_por_pais['country'].tolist()
capacidad_por_tipo_combustible = capacidad_por_tipo_combustible[capacidad_por_tipo_combustible['country'].isin(paises_ordenados)]

# Crear un grÃ¡fico de barras apiladas con los datos filtrados
fig = px.bar(capacidad_por_tipo_combustible, x='country', y='capacity_mw', color='primary_fuel',
             title='Capacidad de GeneraciÃ³n de EnergÃ­a por Tipo de Combustible y PaÃ­s en AmÃ©rica (Ordenado)',
             labels={'capacity_mw': 'Capacidad Total (MW)', 'country': 'PaÃ­s', 'primary_fuel': 'Tipo de Combustible'},
             barmode='stack')

# Ajustar el tamaÃ±o del grÃ¡fico
fig.update_layout(width=1200, height=1000)

# Mostrar el grÃ¡fico interactivo
fig.show()
```

```python
from sklearn.impute import SimpleImputer
df_imputacion = df_americano.copy()
columnas_a_eliminar = ["country",'name', 'gppd_idnr', 'owner', 'url',
                       'geolocation_source', 'wepp_id',
                       'year_of_capacity_data', 'generation_data_source']

df_imputacion = df_imputacion.drop(columns=columnas_a_eliminar)

# Separar variables numÃ©ricas y categÃ³ricas
numericas = ['capacity_mw', 'latitude', 'longitude', 'commissioning_year',
              'generation_gwh_2013', 'generation_gwh_2014',
             'generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017',
             'generation_gwh_2018', 'generation_gwh_2019', 'estimated_generation_gwh_2013',
             'estimated_generation_gwh_2014', 'estimated_generation_gwh_2015',
             'estimated_generation_gwh_2016', 'estimated_generation_gwh_2017']

categoricas = ['country_long', 'primary_fuel', 'source','estimated_generation_note_2013',
               'estimated_generation_note_2014', 'estimated_generation_note_2015',
               'estimated_generation_note_2016', 'estimated_generation_note_2017']

# ImputaciÃ³n para variables numÃ©ricas: imputar la media por paÃ­s
for col in numericas:
    df_imputacion[col] = df_imputacion.groupby('country_long')[col].transform(lambda x: x.fillna(x.mean()))

# ImputaciÃ³n para variables categÃ³ricas: imputar el valor mÃ¡s frecuente
imputer_categoricas = SimpleImputer(strategy='most_frequent')
df_imputacion[categoricas] = imputer_categoricas.fit_transform(df_imputacion[categoricas])

# Verificar si quedan valores nulos
print(df_imputacion.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

# Separar variables numÃ©ricas y categÃ³ricas
numericas = ['capacity_mw', 'latitude', 'longitude', 'commissioning_year',
              'generation_gwh_2013', 'generation_gwh_2014',
             'generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017',
             'generation_gwh_2018', 'generation_gwh_2019', 'estimated_generation_gwh_2013',
             'estimated_generation_gwh_2014', 'estimated_generation_gwh_2015',
             'estimated_generation_gwh_2016', 'estimated_generation_gwh_2017']

categoricas = ['country_long', 'primary_fuel', 'source','estimated_generation_note_2013',
               'estimated_generation_note_2014', 'estimated_generation_note_2015',
               'estimated_generation_note_2016', 'estimated_generation_note_2017']

# ImputaciÃ³n para variables numÃ©ricas: imputar la media
imputer_numericas = SimpleImputer(strategy='mean')
df_imputacion[numericas] = imputer_numericas.fit_transform(df_imputacion[numericas])

# ImputaciÃ³n para variables categÃ³ricas: imputar el valor mÃ¡s frecuente
imputer_categoricas = SimpleImputer(strategy='most_frequent')
df_imputacion[categoricas] = imputer_categoricas.fit_transform(df_imputacion[categoricas])

# Verificar si quedan valores nulos
print(df_imputacion.isna().sum())
```

```python
df_imputacion
```

```python
from sklearn.preprocessing import OneHotEncoder

# SelecciÃ³n de variables categÃ³ricas para codificaciÃ³n one-hot
categoricas = ['country_long', 'primary_fuel', 'source','estimated_generation_note_2013',
               'estimated_generation_note_2014', 'estimated_generation_note_2015',
               'estimated_generation_note_2016', 'estimated_generation_note_2017']

# CodificaciÃ³n one-hot
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_features = encoder.fit_transform(df_imputacion[categoricas])

# Crear un DataFrame con las caracterÃ­sticas codificadas
encoded_df_imputacion = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categoricas))

# Eliminar las columnas originales de las variables categÃ³ricas del DataFrame original
df_imputacion = df_imputacion.drop(columns=categoricas)

# Concatenar el DataFrame original con el DataFrame de caracterÃ­sticas codificadas
df_imputacion = pd.concat([df_imputacion.reset_index(drop=True), encoded_df_imputacion], axis=1)

# Verificar el DataFrame resultante
print(df_imputacion.head())
```

```python
from sklearn.model_selection import train_test_split

# Seleccionar las caracterÃ­sticas para entrenamiento y el objetivo de la predicciÃ³n
caracteristicas = df_imputacion.drop(columns=['generation_gwh_2013', 'generation_gwh_2014',
                                   'generation_gwh_2015', 'generation_gwh_2016',
                                   'generation_gwh_2017', 'generation_gwh_2018',
                                   'generation_gwh_2019'])
objetivo = df_imputacion['generation_gwh_2013']  # Por ejemplo, elegimos el aÃ±o 2013 como objetivo de predicciÃ³n

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(caracteristicas, objetivo, test_size=0.3, random_state=42)

# Ver las dimensiones de los conjuntos de datos
print("Dimensiones del conjunto de entrenamiento:", X_train.shape)
print("Dimensiones del conjunto de prueba:", X_test.shape)
```

PROBANDO LOS ESTIMADORES DE REGRESION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Inicializar el modelo de regresiÃ³n lineal
modelo_regresion_lineal = LinearRegression()

# Entrenar el modelo de regresiÃ³n lineal
modelo_regresion_lineal.fit(X_train, y_train)

# Predecir la generaciÃ³n de energÃ­a en el conjunto de entrenamiento
predicciones_entrenamiento = modelo_regresion_lineal.predict(X_train)

# Evaluar el modelo en el conjunto de entrenamiento
mse_entrenamiento = mean_squared_error(y_train, predicciones_entrenamiento)
r2_entrenamiento = r2_score(y_train, predicciones_entrenamiento)

print("MSE en el conjunto de entrenamiento:", mse_entrenamiento)
print("R^2 en el conjunto de entrenamiento:", r2_entrenamiento)
```

```python
# Predecir la generaciÃ³n de energÃ­a en el conjunto de prueba
predicciones_prueba = modelo_regresion_lineal.predict(X_test)

# Evaluar el modelo en el conjunto de prueba
mse_prueba = mean_squared_error(y_test, predicciones_prueba)
r2_prueba = r2_score(y_test, predicciones_prueba)

print("MSE en el conjunto de prueba:", mse_prueba)
print("R^2 en el conjunto de prueba:", r2_prueba)
```

```python
from sklearn.linear_model import Ridge

# Definir el valor del parÃ¡metro de regularizaciÃ³n (alfa)
alfa = 1.0

# Inicializar el modelo de regresiÃ³n Ridge
modelo_ridge = Ridge(alpha=alfa)

# Entrenar el modelo de regresiÃ³n Ridge
modelo_ridge.fit(X_train, y_train)

# Predecir la generaciÃ³n de energÃ­a en el conjunto de prueba
predicciones_prueba_ridge = modelo_ridge.predict(X_test)

# Evaluar el modelo de regresiÃ³n Ridge en el conjunto de prueba
mse_prueba_ridge = mean_squared_error(y_test, predicciones_prueba_ridge)
r2_prueba_ridge = r2_score(y_test, predicciones_prueba_ridge)

print("MSE en el conjunto de prueba (regresiÃ³n Ridge):", mse_prueba_ridge)
print("R^2 en el conjunto de prueba (regresiÃ³n Ridge):", r2_prueba_ridge)
```

```python
from sklearn.linear_model import Lasso

# Definir el valor del parÃ¡metro de regularizaciÃ³n (alfa)
alfa = 1.0

# Definir el nÃºmero mÃ¡ximo de iteraciones
max_iter = 10000  # Este valor puede ajustarse segÃºn sea necesario

# Inicializar el modelo de regresiÃ³n Lasso con el nÃºmero mÃ¡ximo de iteraciones
modelo_lasso = Lasso(alpha=alfa, max_iter=max_iter)

# Entrenar el modelo de regresiÃ³n Lasso
modelo_lasso.fit(X_train, y_train)

# Predecir la generaciÃ³n de energÃ­a en el conjunto de prueba
predicciones_prueba_lasso = modelo_lasso.predict(X_test)

# Evaluar el modelo de regresiÃ³n Lasso en el conjunto de prueba
mse_prueba_lasso = mean_squared_error(y_test, predicciones_prueba_lasso)
r2_prueba_lasso = r2_score(y_test, predicciones_prueba_lasso)

print("MSE en el conjunto de prueba (regresiÃ³n Lasso):", mse_prueba_lasso)
print("R^2 en el conjunto de prueba (regresiÃ³n Lasso):", r2_prueba_lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Inicializar el modelo de Ã¡rbol de decisiÃ³n
modelo_arbol_decision = DecisionTreeRegressor()

# Entrenar el modelo de Ã¡rbol de decisiÃ³n
modelo_arbol_decision.fit(X_train, y_train)

# Predecir la generaciÃ³n de energÃ­a en el conjunto de prueba
predicciones_prueba_arbol_decision = modelo_arbol_decision.predict(X_test)

# Evaluar el modelo de Ã¡rbol de decisiÃ³n en el conjunto de prueba
mse_prueba_arbol_decision = mean_squared_error(y_test, predicciones_prueba_arbol_decision)
r2_prueba_arbol_decision = r2_score(y_test, predicciones_prueba_arbol_decision)

print("MSE en el conjunto de prueba (Ã¡rbol de decisiÃ³n):", mse_prueba_arbol_decision)
print("R^2 en el conjunto de prueba (Ã¡rbol de decisiÃ³n):", r2_prueba_arbol_decision)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Inicializar el modelo de Random Forest
modelo_random_forest = RandomForestRegressor()

# Entrenar el modelo de Random Forest
modelo_random_forest.fit(X_train, y_train)

# Predecir la generaciÃ³n de energÃ­a en el conjunto de prueba
predicciones_prueba_random_forest = modelo_random_forest.predict(X_test)

# Evaluar el modelo de Random Forest en el conjunto de prueba
mse_prueba_random_forest = mean_squared_error(y_test, predicciones_prueba_random_forest)
r2_prueba_random_forest = r2_score(y_test, predicciones_prueba_random_forest)

print("MSE en el conjunto de prueba (Random Forest):", mse_prueba_random_forest)
print("R^2 en el conjunto de prueba (Random Forest):", r2_prueba_random_forest)
```

```python
import xgboost as xgb

# Convertir los conjuntos de datos a la estructura DMatrix requerida por XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Definir los parÃ¡metros del modelo
parametros = {
    'objective': 'reg:squarederror',  # La funciÃ³n de pÃ©rdida para regresiÃ³n
    'eval_metric': 'rmse'             # La mÃ©trica de evaluaciÃ³n (RMSE)
}

# Entrenar el modelo XGBoost
num_round = 1000  # NÃºmero de rondas de entrenamiento
modelo_xgboost = xgb.train(parametros, dtrain, num_round)

# Realizar predicciones en el conjunto de prueba
predicciones_prueba_xgboost = modelo_xgboost.predict(dtest)

# Evaluar el modelo XGBoost en el conjunto de prueba
mse_prueba_xgboost = mean_squared_error(y_test, predicciones_prueba_xgboost)
r2_prueba_xgboost = r2_score(y_test, predicciones_prueba_xgboost)

print("MSE en el conjunto de prueba (XGBoost):", mse_prueba_xgboost)
print("R^2 en el conjunto de prueba (XGBoost):", r2_prueba_xgboost)
```

```python
from sklearn.model_selection import GridSearchCV

# Definir la cuadrÃ­cula de hiperparÃ¡metros
parametros_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300]
}

# Inicializar el modelo XGBoost
modelo_xgboost = xgb.XGBRegressor()

# Inicializar la bÃºsqueda de cuadrÃ­cula
grid_search = GridSearchCV(modelo_xgboost, parametros_grid, cv=5, scoring='r2')

# Ejecutar la bÃºsqueda de cuadrÃ­cula en los datos de entrenamiento
grid_search.fit(X_train, y_train)

# Obtener los mejores hiperparÃ¡metros encontrados
mejores_hiperparametros = grid_search.best_params_

print("Mejores hiperparÃ¡metros encontrados:", mejores_hiperparametros)
```

```python
# Reentrenar el modelo XGBoost con los mejores hiperparÃ¡metros encontrados
modelo_xgboost_optimizado = xgb.XGBRegressor(learning_rate=0.1, max_depth=3, n_estimators=100)
modelo_xgboost_optimizado.fit(X_train, y_train)

# Predecir la generaciÃ³n de energÃ­a en el conjunto de prueba con el modelo optimizado
predicciones_prueba_xgboost_optimizado = modelo_xgboost_optimizado.predict(X_test)

# Evaluar el modelo XGBoost optimizado en el conjunto de prueba
mse_prueba_xgboost_optimizado = mean_squared_error(y_test, predicciones_prueba_xgboost_optimizado)
r2_prueba_xgboost_optimizado = r2_score(y_test, predicciones_prueba_xgboost_optimizado)

print("MSE en el conjunto de prueba (XGBoost optimizado):", mse_prueba_xgboost_optimizado)
print("R^2 en el conjunto de prueba (XGBoost optimizado):", r2_prueba_xgboost_optimizado)
```

```python
import xgboost as xgb

# Crear y entrenar el modelo XGBoost con los mejores hiperparÃ¡metros encontrados
modelo_xgboost_final = xgb.XGBRegressor(learning_rate=0.1, max_depth=3, n_estimators=100)
modelo_xgboost_final.fit(X_train, y_train)

# Realizar predicciones en nuevos datos
predicciones_nuevos_datos = modelo_xgboost_final.predict(X_test)

# Evaluar el modelo XGBoost optimizado en el conjunto de prueba
mse_prueba_xgboost_optimizado = mean_squared_error(y_test, predicciones_prueba_xgboost_optimizado)
r2_prueba_xgboost_optimizado = r2_score(y_test, predicciones_prueba_xgboost_optimizado)
print("MSE en el conjunto de prueba (XGBoost optimizado):", mse_prueba_xgboost_optimizado)
print("R^2 en el conjunto de prueba (XGBoost optimizado):", r2_prueba_xgboost_optimizado)

# Imprimir las predicciones
print(predicciones_nuevos_datos)
```
